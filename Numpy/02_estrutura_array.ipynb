## Case 2 - Criando uma Array e mostrando informações

Você continua analisando os dados da loja.

Agora o gerente quer entender melhor a estrutura dos dados de vendas.

Crie um código com NumPy que:

- crie um array com os valores [12, 15, 9, 18, 20]  

- mostre:  
  - a quantidade total de elementos  
  - o shape  
  - o número de dimensões  
  - o tipo de dado do array  

```python
# Importando o Numpy e criando uma array
import numpy as np

vendas = np.array([12, 15, 9, 18, 20])
```

```python
print("Total de elementos:", vendas.size)   # Mostra a quantidade de valores 
print("Shape:", vendas.shape)               # Mostra o formato, quantas linhas e colunas
print("Número de dimensões:", vendas.ndim)  # Mostra quantas dimensões tem a Array
print("Tipo de dado:", vendas.dtype)        # Mostra tipo do dado armazenado
```
