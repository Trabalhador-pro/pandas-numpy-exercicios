## Case 4 - Aplicando Estatísica basica 

Você está analisando as vendas da semana:

- 12, 15, 9, 18, 20  

A gerente quer respostas rápidas.

Crie um código com NumPy que:

- crie um array com esses valores  

- mostre:  
  - a maior venda  
  - a menor venda  
  - a média das vendas  

```python
# Importando o Numpy e Criando array 1d
import numpy as np

vendas = np.array([12, 15, 9, 18, 20])

vendas
```

```python
print("Maior venda:", np.max(vendas))       # Mostra o maior valor
print("Menor venda:", np.min(vendas))       # Mostra o menor valor
print("Média das vendas:", np.mean(vendas)) # Mostra a média dos valores
```
