## Case 42 - merge de tabelas

Você tem duas tabelas.

**Tabela 1 - clientes**

- id_cliente: 1, 2, 3, 4  
- nome: Ana, Bruno, Carlos, Daniela  

**Tabela 2 - pedidos**

- id_cliente: 1, 2, 2, 4  
- valor_pedido: 120, 250, 180, 300  

Faça um código com pandas que:

- crie os dois DataFrames  
- use merge para juntar as tabelas pela coluna id_cliente  
- mostre o resultado final  

```python
# Importando o pandas e criando um DataFrame
import pandas as pd

tb_1 = pd.DataFrame({
    'id_cliente': [1, 2, 3, 4 ],
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela']
})

tb_2 = pd.DataFrame({
    'id_cliente': [1, 2, 2, 4 ],
    'valor_pedido': [120, 250, 180, 300 ]
})
```

```python
# Unindo as tabelas pela chave comun dentro de id_cliente. Sujiro pesquisar mais sobre, existem varias coisas a se atentar ao usar merge
# Nesse caso Bruno aparece 2 vezes por conter 2 id dele na outra tabela, e o carlos não aparece por não ter id dele 
# Como se fosse o inner joi do SQL 
df_unida = pd.merge(tb_1, tb_2, on='id_cliente')

df_unida
```
