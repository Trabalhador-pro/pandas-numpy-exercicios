## Case 29 - uso de loc e iloc

Use estes dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, SP, RJ, SP  
- valor_pedido: 120, 250, 80, 300, 150  

Faça um código com pandas que:

- crie o DataFrame  
- usando iloc, mostre:  
  - a primeira linha  
- usando loc, mostre:  
  - apenas as linhas onde cidade == 'SP'  
- usando loc, mostre apenas as colunas:  
  - cliente  
  - valor_pedido  
- usando loc, mostre apenas os clientes de SP com suas colunas cliente e valor_pedido  

```python
# Importando o pandas e criando um data frame
import pandas as pd

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150,]
})

clientes
```

```python
# Seleciona a primeira linha do DataFrame pela posição numérica usando iloc.
clientes.iloc[0]
```

```python
# Seleciona todas as linhas, mas mostra apenas as colunas 'cliente' e 'valor_pedido'.
clientes.loc[:, ['cliente', 'valor_pedido']]
```

```python
# Filtra e mostra apenas as linhas em que a cidade é SP.
clientes.loc[clientes['cidade'] == 'SP']
```
