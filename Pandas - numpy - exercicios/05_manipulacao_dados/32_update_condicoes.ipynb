## Case 33 - atualização com múltiplas condições

Use os mesmos dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, SP, RJ, SP  
- valor_pedido: 120, 250, 80, 300, 150  

Faça um código com pandas que:

- crie o DataFrame  
- crie uma coluna prioridade com "Normal" para todos  
- usando loc, altere prioridade para "Alta" apenas para clientes que:  
  - sejam da cidade RJ  
  - e tenham valor_pedido >= 250  
- mostre o DataFrame final  

```python
# Importando o pandas e criando um data frame
import pandas as pd

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150,]
})

# Cria a coluna 'prioridade' com o valor padrão 'Normal' para todos os clientes.
clientes['prioridade'] = 'Normal'

clientes
```

```python
# Atualiza a coluna 'prioridade' para 'Alta' apenas nos clientes da cidade RJ
# que também tenham valor_pedido maior ou igual a 250.
clientes.loc[(clientes['cidade'] == 'RJ') & (clientes['valor_pedido'] >= 250), 'prioridade'] = 'Alta'

clientes
```
