## Case 27 - pivot_table com soma e média

Use estes dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva, Felipe, Gabriel  
- cidade: SP, RJ, SP, RJ, SP, RJ, SP  
- valor_pedido: 120, 250, 80, 300, 150, 500, 220  

Faça um código com pandas + NumPy que:

- crie o DataFrame  
- crie a coluna categoria  
  - "Alto" para valor_pedido >= 200  
  - "Baixo" para os demais  
- use pivot_table para mostrar, por cidade:  
  - a soma de valor_pedido  
  - a média de valor_pedido  

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva', 'Felipe', 'Gabriel'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150, 500, 220 ]
})

# Cria a coluna 'categoria', classificando os clientes em Alto ou Baixo conforme o valor do pedido.
clientes['categoria'] = np.where(clientes['valor_pedido'] >= 200, 'Alto', 'Baixo')
```

```python
# Cria uma tabela dinâmica com a cidade nas linhas,
# mostrando a soma e a média do valor dos pedidos para cada combinação.

tabela = clientes.pivot_table(
    index='cidade',
    values='valor_pedido',
    aggfunc=['sum', 'mean']
).reset_index()

tabela
```
