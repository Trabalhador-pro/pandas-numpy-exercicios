## Case 30 - atualização de valores com loc

Use os mesmos dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, SP, RJ, SP  
- valor_pedido: 120, 250, 80, 300, 150  

Faça um código com pandas que:

- crie o DataFrame  
- usando loc, crie uma coluna chamada status  
- coloque "Normal" para todos os clientes  
- depois, usando loc, altere o status para "VIP" apenas para clientes com valor_pedido >= 200  
- mostre o DataFrame final  

```python
# Importando o pandas e criando um data frame
import pandas as pd

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150,]
})

clientes
```

```python
# Cria a coluna 'status' e preenche todas as linhas com o valor 'Normal'
clientes.loc[:, 'status'] = 'Normal'

# Atualiza a coluna 'status' para 'VIP' apenas nos clientes com valor_pedido maior ou igual a 200.
clientes.loc[clientes['valor_pedido'] >= 200, 'status'] = 'VIP'

clientes
```
