
## Case 37 - remoção de valores faltantes com dropna

Use os mesmos dados originais:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, None, RJ, SP  
- valor_pedido: 120, 250, 80, None, 150  

Faça um código com pandas que:

- crie o DataFrame  
- remova as linhas que tenham qualquer valor faltante usando dropna()  
- mostre o resultado final  

```python
# Importando o pandas e criando um data frame
import pandas as pd

# Criando um DataFrame com alguns valores faltantes
clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', None, 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, None, 150,]
})

clientes
```

```python
# Visualizamos primeiro qual colunas falta dados
clientes.isna().sum()
```

```python
# Usando DropNa para remover linhas com valores faltantes
clientes.dropna()   # ele retorna um novo DataFrame, sem modificar a tabela original
```
