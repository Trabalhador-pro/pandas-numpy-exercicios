## Case 41 - preenchimentos combinados

Use os mesmos dados originais:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, None, RJ, SP  
- valor_pedido: 120, 250, 80, None, 150  

Faça um código com pandas que:

- crie o DataFrame  
- preencha cidade com a moda  
- preencha valor_pedido com a média  
- mostre o DataFrame final sem valores faltantes  

```python
# Importando o pandas e criando um data frame
import pandas as pd

# Criando um DataFrame com alguns valores faltantes
clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', None, 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, None, 150,]
})

clientes
```

```python
# Atribuindo valores para substituir valores null
media = clientes['valor_pedido'].mean() # Pegando a média
moda = clientes['cidade'].mode()[0]     # Em caso de conter 2 valores que mais se repetem o dado gerado vai ser uma serie com [0] no final pega somente a primeira

# Usando o fillna para substituir apenas algumas colunas especificas
# Trocando pelas atribuições desejadas 
clientes['valor_pedido'] = clientes['valor_pedido'].fillna(media)
clientes['cidade'] = clientes['cidade'].fillna(moda)

# Um data set com valores tratados, não apenas retirados
clientes
```
