## Case 38 - remoção condicional de valores faltantes

Use os mesmos dados originais:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, None, RJ, SP  
- valor_pedido: 120, 250, 80, None, 150  

Faça um código com pandas que:

- crie o DataFrame  
- remova apenas as linhas em que valor_pedido está faltando  
- mostre o resultado final  


```python
# Importando o pandas e criando um data frame
import pandas as pd

# Criando um DataFrame com alguns valores faltantes
clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', None, 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, None, 150,]
})

clientes
```

```python
# Msotrando a quantidade de valor faltando em cada coluna
clientes.isna().sum()
```

```python
# Remove apenas as linha em que o dados está faltando em uma determinada coluna
clientes.dropna(subset=['valor_pedido']) # Retorna outra DataFrame
```
