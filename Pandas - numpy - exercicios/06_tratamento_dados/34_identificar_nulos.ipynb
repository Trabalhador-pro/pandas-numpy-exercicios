## Case 34 - identificação de valores faltantes

Crie um código com pandas usando estes dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva  
- cidade: SP, RJ, None, RJ, SP  
- valor_pedido: 120, 250, 80, None, 150  

Faça um código que:

- crie o DataFrame  
- mostre o DataFrame  
- mostre quais valores estão faltando  

```python
# Importando o pandas e criando um data frame
import pandas as pd

# Criando um DataFrame com alguns valores faltantes
clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'cidade': ['SP', 'RJ', None, 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, None, 150,]
})

clientes
```

```python
# Mostra com True e False, onde existem valores faltantes no DataFrame, onde True é onde falta 
clientes.isna()
```
