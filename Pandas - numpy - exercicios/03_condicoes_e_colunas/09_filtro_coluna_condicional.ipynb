## Case 9 - filtro e criação de coluna condicional

Você recebeu o DataFrame abaixo sobre alunos:

- nome: Ana, Bruno, Carlos, Daniela  
- nota: 8.5, 6.0, 7.5, 9.0  

Crie um código com pandas que:

- crie esse DataFrame  

- mostre apenas os alunos com nota maior ou igual a 7  

- crie uma nova coluna chamada aprovado  
  - regra:  
    - True para nota maior ou igual a 7  
    - False para nota menor que 7  

- mostre o DataFrame final  

```python
# Importando o pandas e criando um data frame
import pandas as pd

notas = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'nota': [8.5, 6.0, 7.5, 9.0]
})

notas
```

```python
# Filtrando os alunos com notas maiores ou igual a 7
notas[notas['nota'] >= 7]
```

```python
# Criando uma nova coluna apenas com a mascara booleana
notas['aprovado'] = notas['nota'] >= 7 

notas
```
