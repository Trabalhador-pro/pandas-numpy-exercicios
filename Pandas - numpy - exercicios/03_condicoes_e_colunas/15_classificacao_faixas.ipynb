## Case 15 - classificação em múltiplas faixas

Uma escola quer classificar os alunos em 3 faixas.

Dados:

- nome: Ana, Bruno, Carlos, Daniela, Eva  
- nota: 8.5, 6.0, 7.5, 9.0, 5.5  

Crie um código com pandas + NumPy que:

- crie um DataFrame  

- crie uma coluna chamada faixa com esta regra:  
  - "Alta" para nota maior ou igual a 8  
  - "Média" para nota maior ou igual a 6 e menor que 8  
  - "Baixa" para nota menor que 6  

- mostre o DataFrame final  

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

notas = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'nota': [8.5, 6.0, 7.5, 9.0, 5.5]
})

notas
```

```python
# 'condicao' reúne as regras booleanas para classificar as notas.
# 'valores' define o texto que será atribuído quando cada regra for satisfeita.
# np.select aplica as condições em ordem e cria a coluna 'faixa';
# caso nenhuma condição seja atendida, o valor padrão será 'Baixa'.

condicao = [
    notas['nota'] >= 8,
    notas['nota'] >= 6,
]

valores = ['Alta', 'Média']

notas['faixa'] = np.select(condicao, valores, default='Baixa') # defautl seria o padrao, então tudo que não mudar pelas condições sera baixa

notas
```
