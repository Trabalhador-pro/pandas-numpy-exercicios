## Case 11 - cálculo de colunas e filtro

Você trabalha com dados de funcionários e recebeu esta tabela:

- nome: Ana, Bruno, Carlos, Daniela  
- salario: 2500, 1800, 3200, 2100  

Crie um código com pandas que:

- crie um DataFrame com esses dados  

- crie uma nova coluna chamada bonus  
  - regra: 10% do salário  

- crie uma nova coluna chamada salario_total  
  - regra: salario + bonus  

- mostre apenas os funcionários com salario_total maior que 2500  

```python
# Importando o pandas e criando um data frame
import pandas as pd

funcionario = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'salario': [2500, 1800, 3200, 2100]
})

funcionario
```

```python
# Criando as novas colunas bonus e salario_total
funcionario['bonus'] = funcionario['salario'] * 0.10
funcionario['salario_total'] = funcionario['salario'] + funcionario['bonus']

funcionario
```

```python
# Filtrando apenas funcionários que recebam > 2500
funcionario[funcionario['salario_total'] > 2500]
```
