## Case 10 - média e coluna condicional

Você recebeu dados de temperatura de uma semana:

- dia: Seg, Ter, Qua, Qui, Sex  
- temperatura: 30, 28, 31, 27, 29  

Crie um código com pandas que:

- crie um DataFrame com esses dados  

- mostre apenas os dias com temperatura maior que a média  

- mostre o valor da média da temperatura  

- crie uma nova coluna chamada acima_da_media  
  - com True para temperaturas maiores que a média  
  - e False para as demais  

```python
# Importando o pandas e criando um data frame
import pandas as pd

df_temp = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, 28, 31, 27, 29]
})

df_temp
```

```python
# Filtrando valores maiores que a média
df_temp[df_temp['temperatura'] > df_temp['temperatura'].mean()]
```

```python
# mostrando a média 
df_temp['temperatura'].mean()
```

```python
# Criando uma nova coluna apenas com a mascara booleana
df_temp['acima_da_media'] = df_temp['temperatura'] > df_temp['temperatura'].mean()

df_temp
```
