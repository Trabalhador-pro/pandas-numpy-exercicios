## Case 12 - coluna categórica e filtro

Uma escola quer classificar o desempenho dos alunos.

Você recebeu os dados:

- nome: Ana, Bruno, Carlos, Daniela, Eva  
- nota: 8.5, 6.0, 7.5, 9.0, 5.5  

Crie um código com pandas que:

- crie um DataFrame  

- crie uma coluna chamada situacao  
  - "Aprovado" para nota maior ou igual a 7  
  - "Reprovado" para nota menor que 7  

- mostre apenas os alunos reprovados  

```python
# Importando o pandas e criando um data frame
import pandas as pd
import numpy as np

notas = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva'],
    'nota': [8.5, 6.0, 7.5, 9.0, 5.5]
})

notas
```

```python
# Criando uma coluna com condição com np.where
notas['situacao'] = np.where(notas['nota'] >= 7, 'aprovado', 'reprovado')

notas
```

```python
# Criando um filtro onde mostre apenas os alunos reprovado
notas[notas['situacao'] == 'reprovado']
```
