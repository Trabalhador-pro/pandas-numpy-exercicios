## Case 7 - filtro e média em DataFrame

Você recebeu uma tabela com vendas de uma loja:

- produto: Notebook, Mouse, Teclado, Monitor  
- vendas: 12, 50, 30, 20  

Crie um código com pandas que:

- crie um DataFrame com essas informações  

- mostre apenas os produtos com vendas maiores que 20  

- mostre a média da coluna vendas  |

```python
# Importando o pandas e criando um data frame
import pandas as pd

venda = pd.DataFrame({
    'produto': ["Notebook", "Mouse", "Teclado", "Monitor"],
    'vendas': [12, 50, 30, 20] 
})

venda
```

```python
# Filtrando os produtos com mais de 20 vendas
venda[venda['vendas'] > 20]
```

```python
# Somando a média da coluna vendas 
venda['vendas'].mean()
```
