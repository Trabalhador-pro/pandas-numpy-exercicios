## Case 6 - seleção de coluna e linha em DataFrame

Uma loja tem a seguinte tabela de produtos:

- produto: Notebook, Mouse, Teclado, Monitor  
- preço: 3500, 80, 150, 900  

Crie um código com pandas que:

- crie um DataFrame com essas informações  

- mostre o DataFrame inteiro  

- mostre apenas a coluna preço  

- mostre apenas a linha do produto Monitor  

```python
# Importando o pandas e criando um data frame
import pandas as pd

produtos = pd.DataFrame({
    'produtos': ['Notebook', 'Mouse', 'Teclado', 'Monitor'],
    'preco': [3500, 80, 150, 900]
})

produtos
```

```python
# Mostrando apenas uma coluna
produtos['preco']
```

```python
# seleção por posição independente do rotulo 
produtos.iloc[3]
```
