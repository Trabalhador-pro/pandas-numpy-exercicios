## Case 8 - criação de coluna calculada (faturamento)

Você recebeu dados de faturamento de uma loja:

- produto: Notebook, Mouse, Teclado, Monitor  
- preco: 3500, 80, 150, 900  
- quantidade_vendida: 3, 10, 5, 2  

Crie um código com pandas que:

- crie um DataFrame com esses dados  

- crie uma nova coluna chamada faturamento  
  - usando: preco * quantidade_vendida  

- mostre o DataFrame final  

```python
# Importando o pandas e criando um data frame
import pandas as pd

venda = pd.DataFrame({
    'produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor'],
    'preco': [3500, 80, 150, 900],
    'quantidade_vendida': [3, 10, 5, 2]
})

venda
```

```python
# Criando uma nova coluna Faturamento apartir de outras colunas
venda['faturamento'] = venda['preco'] * venda['quantidade_vendida']

venda
```
