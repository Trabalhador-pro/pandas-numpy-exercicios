## Case 3 - Criando uma Array 2D


Agora a loja quer separar os dados por duas semanas.

As vendas foram:

- Semana 1: 12, 15, 9, 18, 20  
- Semana 2: 14, 11, 13, 17, 19  

Crie um código com NumPy que:

- monte um array 2D com essas duas semanas  

- mostre o array  
- mostre o shape  
- mostre o size  
- mostre o ndim  

```python
# Importando o Numpy e Criando array 2d
import numpy as np

vendas = np.array([[12, 15, 9, 18, 20],
                   [14, 11, 13, 17, 19]])

vendas      
```

```python
print("Total de elementos:", vendas.size)   # Mostra a quantidade de valores 
print("Shape:", vendas.shape)               # Mostra o formato, quantas linhas e colunas
print("Número de dimensões:", vendas.ndim)  # Mostra quantas dimensões tem a Array
```
