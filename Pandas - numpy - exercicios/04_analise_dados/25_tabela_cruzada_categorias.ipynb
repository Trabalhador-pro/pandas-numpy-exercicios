## Case 25 - tabela cruzada de categorias por cidade

Use os mesmos dados e faça um código que:

- crie o DataFrame  
- crie a coluna categoria  
- mostre uma tabela com:  
  - linhas = cidade  
  - colunas = categorias (Alto, Baixo)  
  - valores = quantidade de clientes  
- garanta que, se uma categoria não existir em uma cidade, apareça 0  

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva', 'Felipe', 'Gabriel'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150, 500, 220 ]
})

# Cria a coluna 'categoria', classificando os clientes em Alto ou Baixo conforme o valor do pedido.
clientes['categoria'] = np.where(clientes['valor_pedido'] >= 200, 'Alto', 'Baixo')

clientes
```

```python
# Agrupa por cidade e categoria, conta a quantidade de clientes em cada combinação
# e usa unstack(fill_value=0) para mostrar também as categorias ausentes com valor zero.
tabela = clientes.groupby(['cidade', 'categoria']).size().unstack(fill_value=0).reset_index()
tabela.columns.name = None
tabela
```
