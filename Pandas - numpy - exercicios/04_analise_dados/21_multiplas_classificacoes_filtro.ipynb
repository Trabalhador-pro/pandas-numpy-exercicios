## Case 21 - múltiplas classificações e filtro combinado

Agora vamos misturar duas classificações e depois cruzar isso em um filtro.

Você recebeu os dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva, Felipe  
- valor_pedido: 120, 250, 80, 300, 150, 500  
- quantidade_itens: 2, 5, 1, 6, 3, 8  

Faça um código com pandas + NumPy que:

1. crie um DataFrame  

2. crie uma coluna chamada ticket_medio_item  

3. crie uma coluna chamada perfil_valor  
   - "Alto" para valor_pedido >= 200  
   - "Baixo" para os demais  

4. crie uma coluna chamada perfil_volume  
   - "Grande" para quantidade_itens >= 5  
   - "Pequeno" para os demais  

5. mostre apenas os clientes que tenham:  
   - perfil_valor == "Alto"  
   - perfil_volume == "Grande"  

6. ordene esse resultado por ticket_medio_item, do maior para o menor  

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

vendas = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva', 'Felipe'],
    'valor_pedido': [120, 250, 80, 300, 150, 500],
    'quantidade_itens': [2, 5, 1, 6, 3, 8]
})

vendas
```

```python
# Calcula o ticket médio por item dividindo o valor do pedido pela quantidade de itens.
vendas['ticket_medio_item'] = vendas['valor_pedido'] / vendas['quantidade_itens']

# Classifica o cliente em perfil de valor: Alto para pedidos a partir de 200, senão Baixo.
vendas['perfil_valor'] = np.where(vendas['valor_pedido'] >= 200, 'Alto', 'Baixo')

# Classifica o cliente em perfil de volume: Grande para 5 itens ou mais, senão Pequeno.
vendas['perfil_volume'] = np.where(vendas['quantidade_itens'] >= 5, 'Grande', 'Pequeno')

# Filtra apenas clientes com perfil de valor Alto e perfil de volume Grande,
# depois ordena pelo ticket médio por item do maior para o menor.
vendas[(vendas['perfil_valor'] == 'Alto') & (vendas['perfil_volume'] == 'Grande')].sort_values(by='ticket_medio_item', ascending=False)
```
