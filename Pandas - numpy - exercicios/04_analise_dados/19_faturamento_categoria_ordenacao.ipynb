## Case 19 - faturamento, categorização e ordenação

Uma loja quer analisar melhor seus produtos.

Você recebeu os dados:

- produto: Notebook, Mouse, Teclado, Monitor, Webcam  
- preco: 3500, 80, 150, 900, 250  
- quantidade_vendida: 3, 10, 5, 2, 8  

Faça um código com pandas + NumPy que:

- crie um DataFrame  

- crie uma coluna chamada faturamento  
  - usando preco * quantidade_vendida  

- crie uma coluna chamada categoria  
  - "Alta" para faturamento maior ou igual a 1000  
  - "Baixa" para faturamento menor que 1000  

- mostre apenas os produtos com categoria Alta  

- mostre esses produtos ordenados do maior faturamento para o menor  

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

produtos = pd.DataFrame({
    'produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Webcam'],
    'preco': [3500, 80, 150, 900, 250],
    'quantidade_vendida': [3, 10, 5, 2, 8]
})

produtos
```

```python
# Calcula o faturamento de cada produto multiplicando o preço pela quantidade vendida.
# Depois classifica os produtos em categoria Alta ou Baixa com base no faturamento.
# Por fim, filtra apenas os produtos da categoria Alta do maior para o menor.

produtos['faturamento'] = produtos['preco'] * produtos['quantidade_vendida']
produtos['categoria'] = np.where(produtos['faturamento'] >= 1000, 'Alta', 'Baixa')

produtos[produtos['categoria'] == 'Alta'].sort_values(by='faturamento', ascending=False)
```
