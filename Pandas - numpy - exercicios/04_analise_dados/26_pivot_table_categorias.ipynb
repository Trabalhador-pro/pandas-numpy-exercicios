## Case 26 - pivot_table com contagem por categoria

Use os mesmos dados:

- cliente: Ana, Bruno, Carlos, Daniela, Eva, Felipe, Gabriel  
- cidade: SP, RJ, SP, RJ, SP, RJ, SP  
- valor_pedido: 120, 250, 80, 300, 150, 500, 220  

Faça um código com pandas + NumPy que:

- crie um DataFrame  
- crie a coluna categoria  
  - "Alto" para valor_pedido >= 200  
  - "Baixo" para os demais  
- use pivot_table para montar uma tabela com:  
  - linhas = cidade  
  - colunas = categoria  
  - valores = quantidade de clientes  
- garanta que valores ausentes apareçam como 0   

```python
# Importando o pandas, Numpy e criando um data frame
import pandas as pd
import numpy as np

clientes = pd.DataFrame({
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eva', 'Felipe', 'Gabriel'],
    'cidade': ['SP', 'RJ', 'SP', 'RJ', 'SP', 'RJ', 'SP'],
    'valor_pedido': [120, 250, 80, 300, 150, 500, 220 ]
})

# Cria a coluna 'categoria', classificando os clientes em Alto ou Baixo conforme o valor do pedido.
clientes['categoria'] = np.where(clientes['valor_pedido'] >= 200, 'Alto', 'Baixo')

clientes
```

```python
# Cria uma tabela dinâmica com as cidades nas linhas e as categorias nas colunas,
# contando quantos clientes existem em cada combinação.
tabela = clientes.pivot_table(
    index='cidade',           # Cria as linhas
    columns='categoria',      # cria as colunas  
    aggfunc='size',           # As funções que serão agregadas  
    fill_value=0              # garante que combinações ausentes apareçam como zero  
).reset_index()

tabela.columns.name = None    # limpa a coluna no indice  
tabela
```
